```python
# Boxplot
plt.figure()
sns.boxplot(y=car_auction['VehOdo'])
plt.title('Boxplot of Vehicle Odometer (VehOdo)')
plt.ylabel('Miles')
plt.show()
```

```python
# Barplot
plt.figure()
sns.countplot(x='Size', data=car_auction)
plt.title('Count of Vehicles by Size')
plt.xlabel('Size')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure()
sns.histplot(car_auction['WarrantyCost'].dropna(), bins=40, kde=False)
plt.title('Distribution of WarrantyCost')
plt.xlabel('WarrantyCost')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Pairplot excluding the 'over_70k_miles' column
cols_no_flag = ['Auction', 'Color', 'IsBadBuy', 'MMRCurrentAuctionAveragePrice', 'Size',
       'TopThreeAmericanName', 'VehBCost', 'VehicleAge', 'VehOdo',
       'WarrantyCost', 'WheelType']
sns.pairplot(car_auction[cols_no_flag], diag_kind='hist')
plt.show()
```

```python
#Pairplot with hue set for IsBadBuy (excluding over_70k_miles)
sns.pairplot(car_auction[cols_no_flag], hue='IsBadBuy', diag_kind='hist')
plt.show()
```
