```python
# TASK II : Changing existing code blocks

import matplotlib.pyplot as plt
import numpy as np

#range of data
x=np.arange(100)
y=np.random.normal(200,2,100)

#Create plots
fig, ax=plt.subplots()
ax.plot(x,y)

#Color,Transparency edit, alpha value
ymin=ax.get_ylim()[0]
ax.fill_between(x, y, ymin, facecolor='y', alpha=.45)

#Title, label
plt.title("Cost over time")
plt.xlabel("time")
plt.ylabel("cost")

#display.display(fig)여
plt.show()
```

I modified the graph in the following ways:
* I changed the background color of the area under the line to **yellow** using `facecolor='y'`.
*  I set the transparency using `alpha=0.45`, which controls how see-through the background color appears. Lower alpha means more transparent, and higher alpha makes it more solid.
*  I updated the graph title to **"Cost over time"**.
* I added axis labels: **"time"** for the x-axis and **"cost"** for the y-axis to make the graph easier to interpret.

Axis labels are important because they show what each part of the graph represents. </br>
The alpha parameter is useful for visual clarity — it helps you balance visibility when overlaying elements. </br>
When I tried changing the alpha value, the yellow area became lighter or darker depending on the number.

```python
#TASK V : Explore your dataframe and data

import pandas as pd

df =pd.read_csv((url))

n_rows, n_cols = df.shape
print(f"[Shape] rows={n_rows}, cols={n_cols}\n")
# DataFrame contains 8 rows and 3 columns.

print("[df.info()]")
df.info()
print()

categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
numeric_cols      = df.select_dtypes(include=["number"]).columns.tolist()

print(f"[Types] categorical/object columns ({len(categorical_cols)}): {categorical_cols}")
print(f"[Types] numeric (int/float) columns ({len(numeric_cols)}): {numeric_cols}\n")

nulls_by_col = df.isna().sum()
cols_with_nulls = nulls_by_col[nulls_by_col > 0]
rows_with_nulls = df.isna().any(axis=1).sum()

print("[Nulls] nulls per column:")
print(nulls_by_col.to_string())
print()
print(f"[Nulls] columns with ≥1 null: {len(cols_with_nulls)}")
print(f"[Nulls] rows with ≥1 null: {rows_with_nulls}")
```

* DataFrame contains 8 rows and 3 columns.

The dataframe has **3 columns**:

* **2 numeric (int/float):** `customer_id`, `sale_amount` </br>
* **1 categorical (object):** `category`

There are **3 null (NaN) values** in total:

* `customer_id`: 1 missing </br>
* `sale_amount`: 1 missing </br>
* `category`: 2 missing </br>

A **NaN (null) value** means missing data. They are important because they can affect calculations, summaries, and analysis accuracy, so we must handle them properly.
